```
print(len(feeds))
errors = []
for i in range(len(feeds)):
    try:
        print(feeds[i]['feed']['title'])
    except:
        print('error en feed: ' + str(i))
        # Almacena indices de feeds que no tienen campo Title
        # y que probablemente ya no existen
        errors.append(i)

print (errors)
print ("\n")
```

```
# Elimina feeds que dan error por no tener title
for j in range(len(errors)):
    feeds.remove(feeds[errors[j]-j])

for feed in feeds:
    # Imprime nombre del RSS Feed
    print("--- " + feed['feed']['title'] + " ---" + "\n")
    for post in feed.entries:
        try:
            # Imprime detalles de los POSTS en los Feeds
            print ("title:   " + post['title'])
            # Elimina con REGEX algunos tags HTML devueltos en summary
            print ("summary: " + re.sub('<.*?>|\n|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});', '', post['summary']))
            print ("link:    " + post['link'])
            print ("updated: " + post['updated'])
            print ("\n" )
        except:
            pass
```

```
data = []
for feed in feeds:
    for post in feed.entries:
        try:
            data.append([feed['feed']['title'],
                         post['title'],
                         re.sub('<.*?>|\n|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});', '', post['summary']),
                         post['link'],
                         post['updated']])
        except:
            pass
df2 = pd.DataFrame(data, columns = ['RSS Feed','Post','Summary','Link','Updated'])
```
